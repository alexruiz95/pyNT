```python
import sympy as sp
import numpy as np
from ode import ode
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
p = sp.Symbol('p')
q = sp.Symbol('q')

A = 1.
B = 1.
C = 0.5

def Usp(x, y):
    return A*sp.cos(x) + B*sp.cos(y) + C*sp.cos(x)*sp.cos(y)
def Unp(x, y):
    return A*np.cos(x) + B*np.cos(y) + C*np.cos(x)*np.cos(y)

H = (p**2 + q**2)/2 + Usp(x, y)

trig_sys = ode(
    x = [x, y, p, q],
    f = [H.diff(p),
         H.diff(q),
         -H.diff(x),
         -H.diff(y)])
```

```python
%matplotlib inline
import matplotlib
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([.1, .1, .8, .8], projection='3d')
X, Y = np.ogrid[.0:3*np.pi:64j, .0:3*np.pi:64j]
ax.plot_wireframe(X, Y, Unp(X, Y))
```

```python
traj = trig_sys.cRK(
    0.01,
    2**14,
    np.array([[3.14, 3.14 ],
              [3.14, 3.14 ],
              [0.90, 0.901],
              [1.61, 1.61 ]]))
```

```python
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111)
ax.plot(traj[:, 0, 0], traj[:, 1, 0])
ax.plot(traj[:, 0, 1], traj[:, 1, 1])
ax.set_aspect('equal')
```

```python
evdt = trig_sys.get_evdt(X0 = np.array([3.14, 3.14, 0.9, 1.61]), solver = ['cRK', 'Taylor4'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(evdt[:, 0], evdt[:, 2])
ax.set_xscale('log')
ax.set_yscale('log')
```
